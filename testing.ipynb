```python
class Example():
    def __init__(self,
                 batch_size = 32,
                 epochs = 10,
                 data = None):
        self.batch_size = batch_size
        self.epochs = epochs
        self.data = data
    def fun(self):
        print(f'Batch size of {self.batch_size}, data is {self.data}')
```

```python
please = Example(data=data,**config)
```
