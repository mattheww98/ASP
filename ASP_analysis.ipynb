# Analysis of ASP results

```python
import json
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
with open("loss_stats.json", "r") as f:
    data = json.load(f)
```

```python
train_mae = data['train']
val_mae = data['val']
epochs = data['epoch']
```

```python
xs = epochs
```

```python
plt.plot(xs,train_mae, 'o-',label=r"""$\mathrm{MAE}_\mathrm{train}$""")
plt.plot(xs,val_mae,'s--',label=r"""$\mathrm{MAE}_\mathrm{val}$""")
plt.xlabel('epochs')
plt.legend()
plt.savefig('ASP_plot.svg')
```
